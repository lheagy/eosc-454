```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid


n_points = 500 
# Generate x values
x = np.linspace(0, 1, n_points)

# Create a noisy function: Sum of sine waves + noise
y = np.cos(4*np.pi*x) #+ 0.2*np.sin(10*np.pi*x) #+ 0.2 * np.sin(20*np.pi*x)
y = y + 0.01*np.random.randn(n_points)

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, y)
ax.set_title("original function")
```

```python
# Compute integral (cumulative integration)
def cumulative_trapezoid(y, x):
    integral = np.zeros_like(y)  # Initialize integral array
    for i in range(1, len(y)):  
        dx = x[i] - x[i - 1]
        integral[i] = integral[i - 1] + (y[i] + y[i - 1]) * dx / 2
    return integral

y_integral = cumulative_trapezoid(y, x)

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, y_integral, color="C1")
ax.set_title("integrated function")
```

```python
# Compute derivative (numerical differentiation)
def numerical_derivative(y, x):
    dy_dx = np.zeros_like(y)  # Initialize derivative array
    dx = np.diff(x)  # Differences between x points
    
    dy_dx[1:] = (y[1:] - y[:-1]) / dx
    return dy_dx
    
dy_dx = numerical_derivative(y, x)

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, dy_dx, color="C2")
ax.set_title("differentiated function")
```
