```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 500 

x = np.linspace(0, 1, n_points) 
y_clean = np.cos(4*np.pi*x)
y = y_clean + 0.05 * np.random.randn(n_points) 
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, y, label="noisy data")
ax.plot(x, y_clean, "k", label="clean_data")
ax.legend()
ax.set_title("original data")
```

```python
def cumulative_trapezoid(x, y): 
    integral = np.zeros_like(x)
    
    for i in range(1, len(y)): 
        dx = x[i] - x[i-1]
        integral[i] = integral[i-1] + 1/2 * (y[i] + y[i-1]) * dx
    return integral 
```

```python
y_integral = cumulative_trapezoid(x, y) 

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, y_integral, color="C1")
ax.set_title("integrated function")
```

```python
def numerical_derivative(x, y): 
    dx = np.diff(x) 

    dy_dx = np.zeros_like(x) 

    # forward difference 
    dy_dx[:-1] = (y[1:] - y[:-1]) / dx 

    # handle our last point (backward difference) 
    dy_dx[-1] = (y[-1] - y[-2])/dx[-1]
    return dy_dx
```

```python
dy_dx = numerical_derivative(x, y)

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
ax.plot(x, dy_dx, color="C2")
ax.set_title("differentiated function")
```
